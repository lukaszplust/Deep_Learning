### ReLU

### $f(x) = max(x,0)$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()#ustawia t≈Ço domyslnie z biblioteki seaborn
```

```python
def relu(x):
  return max(x,0)
```

```python
for i in [-10, -5, 0, 5, 10]:
  print(relu(i))
```

Relu plot

```python
data_relu = np.arange(-10, 20, 0.1) # wartosci od -10 do 20 z krokiem 0.1
```

```python
plotting_data_relu = [relu(x) for x in data_relu]
```

```python
plt.plot(data_relu, plotting_data_relu)
plt.title("Function ReLU")
plt.xlabel("Input data")
plt.ylabel("Calculated values")
```

### SIGMOID

### $ f(x) = \frac{1}{1+e^{-x}}$

```python
def sigmoid(x):
  return 1 / (1.0 + np.exp(-x))
```

```python
for i in [-10, -7, -3, -1, 0 ,1, 3, 7, 10]:
  print(sigmoid(i))
```

```python
data_sigmoid = np.arange(-10, 10, 0.1) # wartosci od -10 do 20 z krokiem 0.1
```

```python
plotting_data_sigmoid = [sigmoid(x) for x in data_sigmoid]
```

```python
plt.plot(data_sigmoid, plotting_data_sigmoid)
plt.title("Function Sigmoid")
plt.xlabel("Input data")
plt.ylabel("Calculated values")
```

TAN H

```python
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
for i in [-10, -7, -3, -1, 0 ,1, 3, 7, 10]:
  print(tanh(i))
```

```python
data_tanh = np.arange(-10, 10, 0.1) # wartosci od -10 do 20 z krokiem 0.1
```

```python
plotting_data_tanh = [tanh(x) for x in data_tanh]
```

```python
plt.plot(data_tanh, plotting_data_tanh)
plt.title("Function Tanh")
plt.xlabel("Input data")
plt.ylabel("Calculated values")
```
